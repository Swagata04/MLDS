```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(10, 25)
y = np.sin(x) * np.cos(x)**2
z = np.sin(x)/np.cos(x)**2
```

# Line Plot

```python
plt.figure(1, figsize=(20, 10))
plt.plot(x,y, color ="darkgreen", marker="o", linewidth=3, label='Sin wave')
plt.plot(x,z, color ="darkorange", marker="^", linewidth=3, label='Z wave')
plt.xlabel("Values")
plt.ylabel("Sin(x)*cos(x)**2")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2")
plt.xticks(np.arange(10,25, 0.5))
plt.yticks(np.arange(-1,1, 0.1))
plt.show()
```

# Histogram

```python
plt.figure(1, figsize=(20, 10))
plt.hist(y, bins=20, color="darkgreen")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2 ranges")
plt.show()
```

# Bar Plot

```python
plt.figure(1, figsize=(20, 10))
plt.bar(['Python', 'Java','C++', 'JavaScript'],[9.0, 8.0, 8.5, 9.1])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2 ranges")
plt.show()
```

```python
plt.figure(1, figsize=(20, 10))
plt.barh(['Python', 'Java','C++', 'JavaScript'],[9.0, 8.0, 8.5, 9.1])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2 ranges")
plt.show()
```

# Scatter Plot

```python
plt.figure(1, figsize=(16, 8))
x = np.random.random(200)
y = np.random.random(200)
plt.scatter(x, y, color='Purple', marker='o')
plt.scatter(x*2, y, color='green', marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title("x vs y")
plt.ylim(0.2, 0.8)
plt.xlim(0.25, 1.75)
plt.legend()
plt.show()
```

# 3D Visualization

```python
axes = plt.axes(projection='3d')
z = x*y
axes.scatter3D(x, y, z)
axes.set_xlabel('X')
axes.set_ylabel("Y")
axes.set_zlabel("Z")
plt.show()
plt.savefig('test.jpg')
```
