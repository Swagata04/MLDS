```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(10, 25)
y = np.sin(x) * np.cos(x)**2
```

# Line Plot

```python
plt.figure(1, figsize=(20, 10))
plt.plot(x,y, color ="darkgreen", marker="o", linewidth=3)
plt.xlabel("Values")
plt.ylabel("Sin(x)*cos(x)**2")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2")
plt.xticks(np.arange(10,25, 0.5))
plt.yticks(np.arange(-1,1, 0.1))
plt.show()
```

# Histogram

```python
plt.figure(1, figsize=(20, 10))
plt.hist(y, bins=20, color="darkorange")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("x vs Sin(x)*cos(x)**2 ranges")
plt.show()
```
